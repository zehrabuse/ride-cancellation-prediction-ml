## Purpose of This Notebook
This notebook presents key **business insights** derived from the processed ride-hailing dataset.  
The goal is to identify patterns and factors influencing ride cancellations, uncover opportunities for operational improvements, and provide actionable recommendations for improving **platform reliability, customer satisfaction, and driver efficiency**.

### Insight 1: Ride Cancellation Patterns

The dataset contains **93,000 completed rides** and **37,500 cancelled rides**.  
A large share of these cancellations were **initiated by drivers rather than customers**.  
This suggests that driver-related factors may play an important role in ride cancellations.

**Possible explanations include:**

- Drivers cancelling rides due to **low expected earnings or insufficient profitability**
- **Vehicle-related issues** preventing drivers from completing trips
- **Long pickup distances** or **unfavorable traffic conditions**

This pattern may indicate a **supply-side reliability issue**, which could reduce **customer satisfaction and trust in the platform**.

### Insight 2: Ride Cancellations by Hour

Analysis of ride cancellations by hour shows that the **highest number of cancellations occur at 18:00, followed by 19:00 and 17:00**, indicating that cancellations are more frequent during the evening peak hours.  

Interestingly, the **highest cancellation rate** (proportion of cancellations relative to bookings) occurs around **05:00**. This is likely due to **fewer bookings during early morning hours**, so even a small number of cancellations results in a higher relative rate.  

**Business Interpretation:**

- Evening cancellations may be influenced by **traffic congestion, driver availability, or trip distance**, which increases the likelihood of driver-initiated cancellations.  
- Early morning high cancellation rates are likely an artifact of **low booking volume**, rather than operational issues.  

**Recommendation:**  
- Consider **dynamic driver incentives** or **surge adjustments** during evening peak hours to reduce cancellations.  
- Monitor early morning cancellations, but recognize that the high relative rate may be **driven by small sample size** rather than a systemic problem.

### Insight 3: Peak Hour for Ride Cancellations

Analysis of cancellations by hour and minute shows that cancellations generally occur between **18:13 and 20:17**.  
When dividing the timeframe into **15-minute intervals**, it becomes clear that the **highest concentration of cancellations is around 18:00**, indicating that **18:00 represents the peak hour for ride cancellations**.

**Business Interpretation:**

- Evening peak hour cancellations are likely influenced by **traffic congestion, high demand, and driver availability**, making this period particularly critical for platform reliability.  
- If cancellations are **customer-initiated**, targeted incentives could encourage bookings during these hours.  
- **Loyal customers** can be identified based on **their booking frequency, reviews, and value to the platform**, as well as **driver feedback**.  

**Recommendations:**  

- Introduce **driver incentives** or **dynamic allocation strategies** around 18:00 to ensure adequate supply and reduce cancellations.  
- Provide **loyal customers with discounts or special offers** during peak hours to encourage bookings and reduce customer-initiated cancellations.  
- Consider **bonus or reward programs** for customers who make bookings during high-demand periods.

### Insight 4: Ride Cancellations by Month

Analysis of ride cancellations by month shows that the **highest number of cancellations occur in May and July**.  

**Business Interpretation:**

- The increase in cancellations during these months may be influenced by **seasonal behavior**.  
- During summer, customers might **prefer walking or alternative modes of transportation** instead of taking rides in cars, possibly due to **desire for fresh air or avoiding traffic**.  
- Since the dataset only covers **2024**, it is not possible to make a **year-over-year seasonal comparison**.

**Recommendation:**  

- Consider **seasonal promotions or incentives** during months with high cancellations to encourage ride usage.  
- Further analysis in future years could help **validate seasonal trends** and guide marketing strategies.

### Insight 5: Ride Cancellations by Day of the Week

Analysis shows that ride cancellations occur **more frequently on weekdays** compared to weekends, with **Monday having the highest number of cancellations**.  
Although the numbers are fairly close across weekdays, the trend suggests a **weekend vs weekday pattern**.

**Business Interpretation:**

- Higher cancellations on weekdays may be influenced by **traffic congestion, risk of being late, or delays in ride arrivals**.  
- Monday shows the highest cancellations, likely because it is the **first workday of the week**, when people are more sensitive to delays.  
- From the driver’s perspective, **high demand during weekdays, especially Monday**, may lead them to prioritize trips that offer higher earnings, cancelling less profitable rides.

**Recommendations:**

- Implement **incentives or priority scheduling** for drivers on high-demand weekdays to reduce cancellations.  
- Consider **customer-focused solutions** for Mondays, such as **early booking reminders or loyalty perks** for consistent riders.

### Insight 6: Ride Cancellations by Vehicle Type

Analysis shows that, although cancellation rates are similar across vehicle types, the **highest number of cancellations occur for Go-Sedan vehicles**.  

**Business Interpretation:**

- High cancellations for Go-Sedan vehicles may be due to **more frequent vehicle malfunctions**, leading to **driver-initiated cancellations**.  
- These vehicles may also be **larger or consume more fuel**, making trips **less profitable for longer rides**, which could cause drivers to cancel.  

**Recommendations:**

- Perform **regular maintenance** on vehicles prone to breakdowns to reduce operational issues.  
- Introduce **driver incentives** to encourage completion of trips that might otherwise be canceled.  
- Offer **customer promotions** highlighting benefits of larger vehicles (e.g., **more comfort, space**), which may reduce customer-initiated cancellations.

## Conclusion

This analysis of ride cancellations has highlighted key patterns and potential drivers affecting both customer- and driver-initiated cancellations.  

**Key takeaways:**

- Driver behavior, vehicle type, and peak hours significantly influence cancellations.  
- Customer preferences, seasonal trends, and payment method issues also play an important role.  
- Operational interventions, targeted incentives, and customer-focused promotions can help reduce cancellations and improve platform reliability.  

These insights provide actionable recommendations for **operations, marketing, and platform strategy teams**. Further analysis and continuous monitoring will ensure that the platform can proactively address cancellation risks and enhance overall customer satisfaction.
